### 变量的类型：
- 严格意义上讲，python只有一个类型。
- 标准数据类型一共6中：
    - 数字类型：Number（没有大小限制，只要内存够）
    - 字符串类型：str
    - 列表：list
    - 元组：tuple
    - 字典：Dictionary
    - 集合：set
    
    <p><b>
    在 python 中，strings, tuples, 和 numbers 是不可更改的对象，而 list,dict 等则是可以修改的对象。<br/><br/>
    不可变类型：变量赋值 a=5 后再赋值 a=10，这里实际是新生成一个 int 值对象 10，再让 a 指向它，而 5 被丢弃，不是改变a的值，相当于新生成了a。    <br/><br/>
    可变类型：变量赋值 la=[1,2,3,4] 后再赋值 la[2]=5 则是将 list la 的第三个元素值更改，本身la没有动，只是其内部的一部分值被修改了。<br/>      <br/>
    </b></p>
    
### 数字：
- 二进制：以0b开头的0,1代码。如：age = 0b10010（十进制的18岁）
- 八进制：以0o开头的包含0-7的数字。如：age = 0o22（十进制的18岁）
- 十六进制：以0x开头的包含0-9，a-f的数字。age = 0xff（十进制的255岁）

- 浮点数：3.14159
    - 3. == 3.0
    - .5 == 0.5
	
- 科学计数法：用e/E后面跟整数表示10的指数。如：176.23 == 1.7623e2；0.876 == 8.76e-1
- 复数（一个由实部和虚部构成的数字，虚部用j/J表示）：如：5+3j，3j，	
- 布尔值：True（1）/False（0），可以跟数字直接操作

### 字符串：
- 如何表示：
    - 单引号：' '->单行字符串
    - 双引号：" "->单行字符串
    - 三单引号''' '''->多行字符串
    - 三双引号""" """->多行字符串

**单双引号可以交替使用**
    
	
- 转义字符：
    - 用\来转义
    - 回车换行符：windows:\n;linux:\r\n

- 字符串格式化：
    - 用百分号格式化：
        - %d:此处为一个整数
        - %s:此处为一个字符串。
        print("I am %s"%"放一个字符串（可以是变量也可以是引号字符串）")
			print("I am %s,I am %d years old"%("字符串参数1",数字参数2))
    - 用format函数：（推荐使用）通过{}和: 代替%号，后面用format带参数完成
			print("I am {}".format("参数"))#注意 . 号
			print("I am {1} year old,I love {0},I am {1} year old".format("zhangsan",18))

### 运算符：
- 算数运算符：+ ，- ，* ，/ ，%(取余)，//取商（也叫地板除），**（幂运算）

**python2中 9/4 等于2，python3中 9/4 等于2.25
python没有自增自减**

- 比较或者关系运算符：== ，!= ，> ，< ，>= ，<=

- 赋值运算符：= ，+= ，-= ，*= ，/= ，//= ，%= ，**=

- 逻辑运算符（没有异或运算）：and ，or ，not
- 运算规则：
    and看做乘法 ，or看做加法
    True看做1 ，False看做0
    则逻辑运算就能转换成整数数学运算
    最后结果如果是0则为False ，否则为True

- 位运算：^ ，| ，>> ，<< ，&

- 成员运算：用来检测一个变量是否是另外一个变量的成员
		in：
		not in：

- 身份运算符：用来检测两个变量是否是"同一个"变量（是否是指向同一个内存）
		is：
		is not:


````python
a = 9
b = 9
print(a is b)
a = "I love you"
b = "I love you"
print(a is b)
````

    <> == !=    等于运算符<br/>

    is is not   身份运算符<br/>
    in not in   成员运算符<br/>
    not or and  逻辑运算符<br/>
